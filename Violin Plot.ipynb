# Violin Plot

Importing the ncessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing the dataset

```python
mart = pd.read_excel("D:/supermarket_sales.xlsx")
mart.head()
```

Creating a simple Violin plot

```python
sns.violinplot(data=mart, y='Total')
```

Creating a voilin plot on two categorical variable and one numerical value and use split

```python
sns.violinplot(data=mart, x="Payment", y="Total")
```

```python
sns.violinplot(data=mart, x="Payment", y="Total", hue="Gender")
```

```python
sns.violinplot(data=mart, x="Payment", y="Total", hue="Gender", split=True)
```

Change the box in the violin plot to horizontal line

```python
sns.violinplot(data=mart, x="Payment", y="Total", hue="Gender", split=True, inner="quartile")
```

Draw line of each of the observation in the violine plot

```python
sns.violinplot(data=mart, x="Payment", y="Total", hue="Gender", split=True, inner="stick")
```

Change the amount of smoothing of line using bw attribute (bandwidth)

```python
sns.violinplot(data=mart, x="Payment", y="Total", hue="Gender", split=True, inner="quartile", bw=1.2)
```

Remove the extreme points

```python
sns.violinplot(data=mart, x="Payment", y="Total", hue="Gender", split=True, inner="quartile", cut=0)
```
