# Box Plot

What information does a box plot give when plotted

        It provides the summary of data which include:
              1) mean                5) 1st Quartile
              2) median              6) 3rd Quartile
              3) minimum value       7) Outliers
              4) maximum value
              

Importing the important libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

importing the dataset

mart = pd.read_excel("D:/supermarket_sales.xlsx")
mart.head()

Creating a simple boxplot on one numerical value

```python
sns.boxplot(data=mart, x="Total")
```

```python
sns.boxplot(data=mart, x="Total", width=0.5)
```

```python
sns.boxplot(data=mart, y="Total", width=0.1)
```

Creating a simple boxplot on one numerical value by  a categorical variable

```python
sns.boxplot(data=mart, x="Payment", y="Total", width=0.2)
```

Creating a simple boxplot on one numerical value by two categorical variable using HUE attribute 

```python
sns.boxplot(data=mart, x="Payment", y="Total", hue="Gender", width=0.5)
```

Add mean marker in the boxplot using the SHOWMEANS attribute and change the color using meanprops

```python
sns.boxplot(data=mart, x="Payment", y="Total", hue="Gender", showmeans=True, width=0.5)
```

```python
sns.boxplot(data=mart, x="Payment", y="Total", hue="Gender", showmeans=True, meanprops={"marker":"o",
                                                                                        "markerfacecolor":"white",
                                                                                        "markeredgecolor":"black",
                                                                                        "markersize":10}, width=0.5)
```

Make the boxplot HORIZONTAL 

```python
sns.boxplot(data=mart, x="Total", y="Payment", hue="Gender", showmeans=True, meanprops={"marker":"o",
                                                                                        "markerfacecolor":"white",
                                                                                        "markeredgecolor":"black",
                                                                                        "markersize":10}, width=0.5)
```

Change the palette, linewidth, etc....

```python
sns.boxplot(data=mart, x="Total", y="Payment", hue="Gender",
            showmeans=True, meanprops={"marker":"o","markerfacecolor":"white","markeredgecolor":"black","markersize":10}, 
            palette="CMRmap",width=0.5)
```

```python
sns.boxplot(data=mart, x="Total", y="Payment", hue="Gender",
            showmeans=True, meanprops={"marker":"o","markerfacecolor":"white","markeredgecolor":"black","markersize":10}, 
            palette="CMRmap",linewidth=2, width=0.5)
```
