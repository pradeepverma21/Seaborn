# Creating a countplot

```python
sns.countplot(data=mart, x="Gender")
```

# Create a countplot to get the number of transaction for each of the product line and change the figure size.

```python
sns.countplot(data=mart, x="Product line")
```

```python
plt.figure(figsize=(15,7))
sns.countplot(data=mart, x="Product line")
```

# Make it horizontal bar plot

```python
sns.countplot(data=mart, y="Product line")
```

# Add hue to get the count on two categories i.e. Product line and Gender

```python
plt.figure(figsize=(15,7))
sns.countplot(data=mart, x="Product line", hue="Gender")
```

# Use different color palatte

```python
plt.figure(figsize=(15,7))
sns.countplot(data=mart, x="Product line", hue="Gender", palette="PuBuGn_r")
```

# Changing Style using facecolor, linewidth and edge color 

```python
 plt.figure(figsize=(15,7))
sns.countplot(data=mart, x="Product line",
              facecolor=(1,0,1,0), 
              linewidth=5, 
              edgecolor=sns.color_palette('dark',3))
```
