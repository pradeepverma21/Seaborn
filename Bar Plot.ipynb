Importing the supermarket sales dataset

```python
mart = pd.read_excel("D:/supermarket_sales.xlsx")
```

```python
mart.head()
```

Creating a simple Bar Plot

```python
sns.barplot(data=mart, x="Product line", y="Total")
```

```python
plt.figure(figsize=(15,7))
sns.barplot(data=mart, x="Product line", y="Total")
```

Add a Hue to the barplot

plt.figure(figsize=(15,7))
sns.barplot(data=mart, x="Product line", y="Total", hue="Gender")

Make the graph Horizontal

```python
plt.figure(figsize=(15,7))
sns.barplot(data=mart, x="Total", y="Product line", hue="Gender")
```

Add cap on the error bar

```python
plt.figure(figsize=(15,7))
sns.barplot(data=mart, x="Product line", y="Total", hue="Gender", capsize=0.1)
```

```python
x=mart["Product line"].sort_values()
x.unique()
```

Remove the error using errorbar

```python
plt.figure(figsize=(15,7))
sns.barplot(data=mart, x="Product line", y="Total", hue = "Gender", errorbar=None)
```

Changing the color using the COLOR attribute

```python
plt.figure(figsize=(15,7))
sns.barplot(data=mart, x="Product line", y="Total", hue = "Gender", errorbar=None, color="red")
```

Changing the palette color using PALETTE attribute

```python
plt.figure(figsize=(15,7))
sns.barplot(data=mart, x="Product line", y="Total",hue="Gender", errorbar=None, palette="cividis")
```

Changing the Saturation of the bar 

```python
plt.figure(figsize=(15,7))
sns.barplot(data=mart, x="Product line", y="Total",hue="Gender", errorbar=None, palette="cividis", saturation=0.3)
```

CHANGE DEFAULT AGGREGATION METHOD USING ESTIMATOR

```python
plt.figure(figsize=(15,7))
sns.barplot(data=mart, x="Product line", y="Total",hue="Gender", errorbar=None, palette="cividis",estimator=sum)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(data=mart, x="Product line", y="Total",hue="Gender", errorbar=None, palette="cividis",estimator = np.mean)
```
