# Strip plot

Why do use the strip plot ?

        A strip plot is a graphical data analysing technique for summarising a univariate dataset.

Importing all the necessary library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing the dataset

```python
mart = pd.read_excel("D:/supermarket_sales.xlsx")
mart.head()
```

Creating a simple Strip plot

```python
sns.stripplot(data=mart, y="Total")
```

Expand markers in strip plot using jitter attribute

```python
sns.stripplot(data=mart, y="Total", jitter=0.2) 
```

Draw the lines arount the points using LINEWIDTH

```python
sns.stripplot(data=mart, y="Total", jitter=0.2, linewidth=0.5)
```

Draw strip plot using multiple categorical variables

```python
sns.stripplot(data=mart, x="Payment", y="Total")
```

```python
sns.stripplot(data=mart, x="Payment", y="Total",hue="Gender")
```

Seperate each level of stripplot using Dodge attribute

```python
sns.stripplot(data=mart, x="Payment", y="Total",hue="Gender", dodge=True)
```

Draw the strip on top of violin plot 

```python
sns.stripplot(data=mart, x="Payment", y="Total",jitter=0.2, color="yellow", linewidth=0.3)
sns.violinplot(data=mart, x="Payment", y="Total",jitter=0.2,color="grey")
```

Draw the strip plot on top of boxplot

```python
sns.stripplot(data=mart, x="Payment", y="Total",jitter=0.2, color="orange", linewidth=0.3)
sns.boxplot(data=mart, x="Payment", y="Total",color="grey")
```

Adding colors to the strips in strip plot

```python
palette = ["yellow","green","orange"]
sns.stripplot(data=mart, x="Payment", y="Total",jitter=0.2, color="orange", palette=palette, linewidth=0.3)
```
